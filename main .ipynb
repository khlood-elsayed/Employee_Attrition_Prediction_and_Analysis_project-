```python
data['Distance from Home'] = data['Distance from Home'].fillna(data['Distance from Home'].median())
data['Company Tenure (In Months)'] = data['Company Tenure (In Months)'].fillna(
    data['Years at Company'] * 12)
data['Company Tenure (In Months)'] = data['Company Tenure (In Months)'].fillna(
    data['Company Tenure (In Months)'].median())
```

```python
median=data['Distance from Home'].median()
data['Distance from Home']=data['Distance from Home'].fillna(median)
```

```python
data=data.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
for col in data.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
data_numeric=data.select_dtypes(exclude='object')
for col in data_numeric.columns:
    q1=data[col].quantile(0.25)
    q3=data[col].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    outlier=((data_numeric[col]>high)|(data_numeric[col]<low)).sum()
    total=data_numeric[col].shape[0]
    print(f"Total Outlier in {col} are {outlier}--{round(100*(outlier)/total,2)}%")
    if outlier>0:
        data=data.loc[(data[col]<=high)&(data[col]>=low)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = data.select_dtypes(include=["int64", "float64"]).columns
num_cols = num_cols.drop("Employee ID")
plt.figure(figsize=(20, 15))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 4, i)
    sns.histplot(data[col], kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

```

```python
cat_cols = data.select_dtypes(include=["object"]).columns

plt.figure(figsize=(16, 12))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(4, 4, i)
    data[col].value_counts(normalize=True).plot.bar()
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

cat_cols = data.select_dtypes(include=["object"]).columns

plt.figure(figsize=(16, 12))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(4, 4, i)
    data[col].value_counts().plot.pie(autopct="%1.1f%%", startangle=90)
    plt.title(f"Pie Chart of {col}")
    plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ["Age", "Years at Company", "Monthly Income",
            "Number of Promotions", "Number of Dependents"]

for col in num_cols:
    plt.figure(figsize=(7,5))
    sns.barplot(x="Attrition", y=col, data=data)
    plt.title(f"{col} Distribution vs Attrition")
    plt.xlabel("Attrition")
    plt.ylabel(col)
    plt.show()
```

```python
cat_cols = ["Gender", "Work-Life Balance", "Job Satisfaction",
            "Performance Rating", "Job Level", "Company Size",
            "Remote Work", "Leadership Opportunities", "Innovation Opportunities","Marital Status"]

for col in cat_cols:
    plt.figure(figsize=(7,5))
    sns.countplot(x=col, hue="Attrition", data=data, palette=["red","blue"])
    plt.title(f"{col} vs Attrition")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.legend(title="Attrition", labels=["Left", "Stayed"])
    plt.xticks(rotation=30)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Job Role", hue="Attrition", data=data, palette=["red","blue"])
plt.title("Job Role vs Attrition")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Job Role", y="Monthly Income", data=data)
plt.title("Monthly Income by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Monthly Income")
plt.xticks(rotation=45)
plt.show()

print(data.groupby("Job Role")["Monthly Income"].mean())
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Job Role", hue="Overtime", data=data)
plt.title("Overtime by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

print(data.groupby("Job Role")["Overtime"].value_counts(normalize=True))
```
